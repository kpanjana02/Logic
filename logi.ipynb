```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

sns.set_theme(style="darkgrid")
```

```python
#load the simple dataset
df=pd.read_csv("pass.csv")
print("Dataset:")
print(df.head())

#visualise the data
plt.figure(figsize=(8,5))
plt.scatter(df['Hours'], df['Class'], c=df['Class'], cmap='bwr', edgecolor='k')
plt.xlabel('Hours Studied')
plt.ylabel('Pass (1)/Fail (0)')
plt.title('Hours vs Result')
plt.show()
```

```python
x=df[['Hours']]
y=df[['Class']]


model=LogisticRegression()
model.fit(x,y)

print("Model trained successfully.")
```

```python
#visualising the Sigmoid Curve(Logistic Regression curve)
plt.figure(figsize=(10,6))
plt.scatter(df['Hours'],df['Class'],c=df['Class'],cmap='bwr',label='Data Points')


x_test_range =np.linspace(0,10,100).reshape(-1,1)
y_prob_range=model.predict_proba(x_test_range)[:,1]

plt.plot(x_test_range,y_prob_range,color='green',linewidth=3,label='Logistic Curve')
plt.axhline(y=0.5,color='gray',linestyle='--',label='Decision Boundary(0.5)')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression Sigmoid Curve')
plt.legend()
plt.show()
```

```python
try:
    #get user input
    user_hours=float(input("Enter the study hours (0,-10): "))

    x_user=np.array([[user_hours]])
    #make prediction
    prediction=model.predict(x_user)[0]
    probability=model.predict_proba(x_user)[0][1]
    #output result
    if prediction==1:
        print(f"\nResult:Pass")
        print(f"Confidence: {probability*100:.2f}%")
    else:
        print(f"\nResult: Fail")
        print(f"Confidence: {(1-probability)*100:.2f}%")

except ValueError:
    print("Invalid input.Please enter a numerical value for hours.")

```
